```python
import yfinance as yf
import pandas as pd
import numpy as np

start_date = '2023-02-01'
end_date   = '2025-01-15'
interval   = '1h'

def get_dataset(start_date, end_date, interval):
    data_dict = {}

    for (lev, inv, base) in triplets:
        df_lev = yf.download(
            lev, 
            start=start_date,
            end=end_date,
            interval=interval
        )
        df_lev = df_lev[['Close', 'Volume']].rename(
            columns={'Close': 'Close_Lev', 'Volume': 'Volume_Lev'}
        )
        
        df_inv = yf.download(
            inv, 
            start=start_date,
            end=end_date,
            interval=interval
        )
        df_inv = df_inv[['Close', 'Volume']].rename(
            columns={'Close': 'Close_Inv', 'Volume': 'Volume_Inv'}
        )

        df_base = yf.download(
            base,
            start=start_date,
            end=end_date,
            interval=interval
        )
        df_base = df_base[['Close']].rename(
            columns={'Close': 'Close_Base'}
        )

        df_merged = df_lev.join(df_inv, how='outer').join(df_base, how='outer')

        df_merged['Volume_Lev'] = df_merged['Volume_Lev'].replace(0, np.nan).ffill()
        df_merged['Volume_Inv'] = df_merged['Volume_Inv'].replace(0, np.nan).ffill()

        data_dict[(lev, inv, base)] = df_merged

    return data_dict
```

```python
triplets = [
    ('UPRO', 'SPXU', '^GSPC'),
    ('SSO',  'SDS',  '^GSPC'),
    ('SPY',  'SH',   '^GSPC'),
    ('TQQQ', 'SQQQ', '^IXIC'),
    ('QLD',  'QID',  '^IXIC'),
    ('QQQ',  'PSQ',  '^IXIC'),
    ('UDOW', 'SDOW', '^DJI'),
    ('DDM',  'DXD',  '^DJI'),
    ('DIA',  'DOG',  '^DJI'),
]


start_date = '2023-02-01'
end_date   = '2025-01-15'
interval   = '1h'
data_hour = get_dataset(start_date, end_date, interval)



for p in triplets:
    df = data_hour[p]      
    df.columns = df.columns.droplevel(level=["Ticker"])
    filename = f"./dataset/{p[0]}_{p[1]}_hour.csv"    
    df.reset_index().to_csv(filename, index=False)
```
